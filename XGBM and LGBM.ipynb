```python
clean_df = df.drop(['Name','Cabin','Ticket','PassengerId'], axis = 1)
clean_df.dropna(inplace=True)
```

```python
clean_df
```

```python
#2.	Encode categorical variables using one-hot encoding or label encoding.

for i in clean_df.columns:
    if len(clean_df[i].value_counts()) > 5:
        print(len(clean_df[i].value_counts()))
        print(f'len is One hot encodding has to applied on column: {i}')
```

```python
from sklearn.preprocessing import LabelEncoder

# applying label encoder if the unique values are less than 6
for i in clean_df.columns:
    if len(clean_df[i].value_counts()) < 6:
        clean_df[i] = LabelEncoder().fit_transform(clean_df[i])
    else:
        # if the unique values are more than 6 than it will apply one hot encoding method
        clean_df = pd.get_dummies(clean_df, columns=[i])

clean_df
```

## Building Predictive Models:

1.	Split the preprocessed dataset into training and testing sets.
2.	Choose appropriate evaluation metrics (e.g., accuracy, precision, recall, F1-score) for model evaluation.
3.	Build predictive models using LightGBM and XGBoost algorithms.
4.	Train the models on the training set and evaluate their performance on the testing set.
5.	Use techniques like cross-validation and hyperparameter tuning to optimize model performance.


```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix

# splitting the data into test and train
X = clean_df.drop('Survived', axis=1)
Y = clean_df['Survived']

X_train, X_test, y_train, y_test  = train_test_split(X,Y, test_size=0.30,random_state=10)
```

## Model Building XGBM

```python
from xgboost import XGBClassifier
```

```python
model = XGBClassifier(n_estimators = 300, max_depth = 4)
model.fit(X_train,y_train)
```

```python
# predicting on the model

y_pred = model.predict(X_test)
y_pred_train = model.predict(X_train)
y_pred
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
confusion_matrix(y_test,y_pred)
```

```python
vis = ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred), display_labels=model.classes_,)
vis.plot()
```

```python
#accuracy score for predictiong unseen data
accuracy_score(y_test,y_pred)*100
```

```python
#accuracy score for predicting train data
accuracy_score(y_train,y_pred_train)*100
```

## Applying cross-validation and hyperparameter tuning to optimize model performance

**HYPERPARAMETER TUNING USING GRIDSEARCH CV**

```python
# hyperparameters for XGBM
from sklearn.model_selection import GridSearchCV,KFold
params ={
    'learning_rate': [0.001,0.002,0.003],
    'booster' :['gbtree','gblinear','dart'],
    'max_depth' : [4,5,6],
    'max_leaves' : [0,1],
    'objective' : ['binary:logistic','binary:hinge','binary:logitraw'],
    'n_estimators' : [300,400,500,600]
}
```

```python
params
```

```python
xgbm_model = XGBClassifier(n_jobs = -1)
kfold = KFold(n_splits=5, shuffle=True)
grid_search = GridSearchCV(estimator=xgbm_model,
                          param_grid=params,
                          cv = kfold,
                          verbose=10,
                          scoring='accuracy')
grid_search.fit(X,Y)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

**Hyperparameter for LGBM**

```python
params1 = {}
params1['learning_rate'] = [0.002,0.1,0.3]
params1['boosting_type'] = ['gbdt','dart']
params1['objective'] = ['binary','multiclass']
params1['metric'] = ['binary_logloss','multi_logloss']
params1['num_leaves'] = [0,1,5]
params1['max_depth'] = [5,6,10]
```

```python
params1
```

```python
model_lgbm = LGBMClassifier(n_jobs=-1)
kfold = KFold(n_splits=5, shuffle=True)
grid_search_lgbm = GridSearchCV(estimator=model_lgbm,
                          param_grid=params1,
                          cv = kfold,
                          verbose=10,
                          scoring='accuracy')
grid_search_lgbm.fit(X,Y)
```

```python
print(grid_search_lgbm.best_params_)
print(grid_search_lgbm.best_score_)
```
